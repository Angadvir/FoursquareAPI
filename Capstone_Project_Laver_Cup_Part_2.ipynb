Introduction: 

The "Laver Cup" is a Tennis Tournament organized each year, in the month of September, for a 3 days showdown of the world's top Tennis talents in the world, competing against one another. The "Laver Cup" pits six top European players against six of their counterparts from the rest of the World.
One of the prime reasons to introduce the "Laver Cup" is to diversify, advocate and market the sport of Tennis to the entire world and also to lesser famous Tennis destinations of the world, where the sport is not a top event of the city. 
In a similar vein, the location of the "Laver Cup" rotates between Europe and rest of the world cities, each year. 
The "Laver Cup" was brought to life in 2017 and has been a mega-success since its inception. 
The Cup was inaugurated in Prague, Czech Republic in 2017, it then travelled to the US for the 1st time in 2018, in Chicago and back to Europe (Geneva, Switzerland), in 2019. In summary: 
1. 2017 - Prague, Czech Republic, Europe
2. 2018 - Chicago, Illinois, US
3. 2019 - Geneva, Switzerland, Europe

Business Problem: 

The Laver Cup committee has been tasked to host the next Laver Cup, scheduled in September, 2020 in one of the major US cities. A number of major US cities entered the draw, out of which a total of 6 major US cities are shortlisted, from the random draw. From the 6 major US cities, the Laver Cup committee must choose a city, from the final 6 cities, to host the Cup. The committee must ensure that the city will be ready from an infrastructure stand-point, to ensure the facilities in and around the main stadium are developed to support a flush of world audience. The committee must also ensure the neighborhood around the stadium is conducive to support a strong sports culture and captivate the Tennis audience for an ultimate fan experience. In short, the committee must find a city which has the potential to host a global tennis tournament, which has not yet been explored to facilitate as a prime destination for competitive Tennis tournament(s), with an interest to widen the roots of Tennis is lesser popular Tennis destinations of the world.

Methodology:
To help the choose the next best US city to host "Laver Cup - 2020", the Tennis stadiums of the 4 major Grand Slams will be explored along with the 3 city centers which have hosted the Laver Cups, in the past. 
A master data set of all the 7 major stadiums (4 Grand Slam stadiums + 3 Laver Cup stadiums) will be created to cluster the stadiums and explore the similarity between the 7 stadiums. 

Next, the 6 candidate city stadiums, in the US, will be explored in a dataframe (DF).

To compare the 6 candidate US cities with the 7 major stadiums, a concatenated "Master Tennis Dataset" will be created which will include the frequency of venue categories against each of 13 major city stadiums (6 candidates + 7 historical stadiums).

ML model(s):
A "KMeans" clustering algorithm will be utilized to segregate the 6 candidate city stadiums into different clusters. A pattern based algorithm was chosen here, since the dataset contains historical, static data on which the KMeans is known to perform with acceptable accuracy. 
In order to decide the number of clusters to solve each of the problems/ datasets, 2 optimization techniques were utilized: 
1. The Elbow Method - This computes the "WSS" or 'within-sum-of-squares' error through finding the euclidian distance of each of the data points within a cluster to the center of other clusters. The point (number of clusters) where the "WSS" loss tends to flatten out (or create an elbow point), is the optimal number of clusters to use for the given dataset. 
2. The Silhoutte Method - This method measures the similarity of an object within a cluster with other objects within the same cluster (cohesiveness) and compares it with the dissimilarity of the object with other clusters (separation). The Silhoutte score ranges from -1 to +1. A high value implies an appropriate clustering configuration, while lower values imply that number of clusters are too many or too few (i.e. the object in the cluster could be better classified compared to other clusters in the dataset). In the graph, the highest point (number of clusters with highest value), is considered as the optimal number of clusters.

After running the "KMeans" algorithm on the "Master Tennis Dataset", the candidate states / candidate cities (out of the 6), which are most similar to US Open and Chicago-Laver Cup event (i.e. if fall under the same cluster), were considered the best cities to host the next Laver Cup. 
Again, measurement metric utilized is 'Neighborhood Similarity'. The US Open was taken as the baseline as it is a considered a good attraction point for many sport enthusiasts and Tennis fans alike, being one of the 4 major Tennis destinations of world.

Aggregating the DF by 'Event_Venue' and finding the mean for each of the categories by 'Event_Venue'. 

```python
#Compute an aggregate of venue categories by aggregtaing the 'Event locations':
tennis_events_aggregated = one_hot_df.groupby('Event_Venue')
tennis_events_agg_mean = one_hot_df.groupby('Event_Venue').mean().reset_index()

#print(tennis_events_aggregated.head())

tennis_events_agg_mean.head()
print(tennis_events_agg_mean.shape)
```

Defining a function to obtain "n" most common venues for each of the 'Event_Venues': 

```python
#Obtain the most common venues: 
def get_common_categories(row, num_top_venues):
    category_rows = row[1:]
    category_rows_sorted = category_rows.sort_values(ascending=False)
    
    return category_rows_sorted.index.values[:num_top_venues]
```

Creating a new DF and applying the "get_common_categories" to obtain 'n' most common venue categories for each of the 'Event_Venue':

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Event_Venue']

for ind in np.arange(num_top_venues):
    try: 
        columns.append('{}{} most common venue category'.format(ind+1,indicators[ind]))
    except:
        columns.append('{}th most common venue category'.format(ind+1))

#Create a new dataframe:
tennis_event_common_venues = pd.DataFrame(columns = columns)        
tennis_event_common_venues['Event_Venue'] = tennis_events_agg_mean['Event_Venue']

for i in np.arange(tennis_events_agg_mean.shape[0]):
    tennis_event_common_venues.iloc[i,1:] = get_common_categories(tennis_events_agg_mean.iloc[i,:], num_top_venues)
        
tennis_event_common_venues
```

Running the 'Elbow-Method' and 'Silhoutte' to determine the optimal number of clusters to initiate for classification of 'Event Venue'. 

```python
#Determing Optimal number of clusters using the Silhouette method:
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

tennis_clusters_dataset = tennis_events_agg_mean.drop(columns = ['Event_Venue'], axis = 1)

# k-means determine k
distortions = []
K = range(1,6)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(tennis_clusters_dataset)
    kmeanModel.fit(tennis_clusters_dataset)
    distortions.append(np.sum(np.min(cdist(tennis_clusters_dataset, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / tennis_clusters_dataset.shape[0])

#Plot the elbow:
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The WSS Method showing the optimal k')
plt.show()
#-------------------------------------------------------------------------------------------------------------------------------------------

from sklearn.metrics import silhouette_score

sil = []
kmax = 6
K = np.arange(2,7)
#Dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
    kmeans = KMeans(n_clusters = k, random_state = 0).fit(tennis_clusters_dataset)
    labels = kmeans.labels_
    sil.append(silhouette_score(tennis_clusters_dataset, labels, metric = 'euclidean'))

print(K, sil)
plt.plot(K, sil, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('The Silhouette Method showing the optimal k')
plt.show()
```

Executing the 'KMeans' algorithm to label the 7 Tennis Venues into 2 clusters (2 clusters were chosen due to highest score at 2 clusters using the 'Silhouette' method): 

```python
from sklearn.cluster import KMeans
#Runnning the K-Means Clustering Algorithm to label the 7 Tennis venues in differnt clusters: 
kclusters = 2

tennis_clusters_dataset = tennis_events_agg_mean.drop(columns = ['Event_Venue'], axis = 1)

#Run the K-Means Clustering algorithm: 
K_Means = KMeans(n_clusters = kclusters, random_state = 0).fit(tennis_clusters_dataset)

K_Means.labels_[:]
```

Determing the optimal number of clusters to initate to classify the 6 candidate locations using 'Elbow Method' and 'Silhoutte Method':

```python
#Determing Optimal number of clusters using the WSS / Elbow method:
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

clustering_dataset = one_hot_master_candidate_means.drop(columns = ['state'], axis = 1)

# k-means determine k
distortions = []
K = range(1,6)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(clustering_dataset)
    kmeanModel.fit(clustering_dataset)
    distortions.append(np.sum(np.min(cdist(clustering_dataset, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / clustering_dataset.shape[0])

#Plot the elbow:
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The WSS Method showing the optimal k')
plt.show()
#-------------------------------------------------------------------------------------------------------------------------------------------
#Determing Optimal number of clusters using the Silhoutte method:

from sklearn.metrics import silhouette_score

sil = []
kmax = 5
K = np.arange(2,6)
#Dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
    kmeans = KMeans(n_clusters = k, random_state = 0).fit(clustering_dataset)
    labels = kmeans.labels_
    sil.append(silhouette_score(clustering_dataset, labels, metric = 'euclidean'))

print(K, sil)
plt.plot(K, sil, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
